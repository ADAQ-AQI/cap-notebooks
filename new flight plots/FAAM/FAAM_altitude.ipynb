# Plot altitude profiles of aircraft pollution data

## Setup imports

```python
import matplotlib.pyplot as plt
import prep as pre
```

## Define plotting function

```python
def plot_altitude_profile(x_data, y_data) :

    fig = plt.figure(figsize=(15,15))
    ax  = plt.axes([.2,.15,.75,.7])

    plt.scatter(x_data,y_data,s=200,label='Aircraft')
    plt.xlabel(species_label,fontsize=50,labelpad=10)
    plt.ylabel('Altitude / m',fontsize=50,labelpad=10)

    ax.tick_params(labelsize=30)

    plt.legend(fontsize=30,markerscale=2)

    plt.show()
    #plt.savefig(pre.plotdir+species_key+'_Altitude_Profile.png')
    #plt.close()
```

## Prepare the data and plot for each pollution species

```python
species_dict = pre.species_dictionary()
df = pre.data

for species in species_dict :
    species_key = species
    species_code = species_dict[species]['code']
    species_label = species_dict[species]['label']

    alt_data = df['Altitude / m'][:].values
    conc_data = df[species_code][:].values

    plot_altitude_profile(conc_data, alt_data)
```
